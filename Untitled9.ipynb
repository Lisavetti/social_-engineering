```python
import requests
from urllib.parse import quote_plus
import dns.resolver
```

```python
def check_mx_record(email):
    domain = email.split('@')[1]
    try:
        # Отримуємо MX записи для домену
        mx_records = dns.resolver.resolve(domain, 'MX')
        if mx_records:
            return True
    except Exception as e:
        print(f"Error checking MX for {email}: {e}")
        return False

# Перевіряємо кожну електронну адресу
for email in email_addresses:
    has_mx = check_mx_record(email)
    print(f"Email {email} {'has valid MX records' if has_mx else 'does not have valid MX records'}")

```

```python
def check_email_validity(email):
    # Кодуємо електронну адресу для використання у URL
    query = quote_plus(f'"{email}" site:google.com')
    url = f"https://www.google.com/search?q={query}"

    try:
        response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})

        # Перевіряємо наявність електронної адреси у відповіді (це дуже базова перевірка)
        if email in response.text:
            return True
        else:
            return False
    except Exception as e:
        print(f"Error checking {email}: {e}")
        return False

# Перевіряємо кожну електронну адресу
for email in email_addresses:
    is_valid = check_email_validity(email)
    print(f"Email {email} is {'valid' if is_valid else 'invalid'}")
```
